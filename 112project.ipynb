```python
def Convertir_formats(df):
    df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
    df['DateofHire'] = pd.to_datetime(df['DateofHire'], errors='coerce')
    df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], errors='coerce')
    df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
    df['Absences']= pd.to_numeric(df['Absences'], errors='coerce')
    
    return df

    Convertir_formats(df)
```

```python
def define_data(df):
    print("This is a data cleaning project.")

    print(f'shape of the data: {df.shape} \n')
    print(f'mean valur manquent =>> {df["DateofTermination"].isna().mean() * 100}')
    print(f'diplicated rows {df.duplicated().sum()} \n')
    print(f'missing values in each column:\n {df.isna().sum()} \n')
    print(f'data types in the dataframe:\n {df.dtypes} \n')
define_data(df)
```
